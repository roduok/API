```python
# Dependencies
import json
import requests as req
import pandas as pd
import matplotlib.pyplot as plt
#Load City List
df = pd.read_json("city.list.json", orient='columns')
# Save config information.
endpoint = "http://api.openweathermap.org/data/2.5/weather"
api_key = "fd9c5f98040dd22bdc74c40c46646db7"
units = "imperial"
city_list = df[["name","id"]]
city_list_random=city_list.sample(n=500)
city_list_random.head()
```

```python
#Build partial query URL
params = {
    'units': units,
}
weather_data = []
row_count = 1
url = "http://api.openweathermap.org/data/2.5/weather?"

# Loop through the list of cities and perform a request for data on each
for index, row in city_list_random.iterrows():
     # Create endpoint URL
    target_url = url + "appid=" + api_key + "&id=" +  str(row["id"])
    print("Now retrieving city # " + str(row_count))
    print(row["name"])
    print(target_url)
    row_count += 1
    #response = req.get(endpoint, params=params).json()
    response = req.get(target_url, params=params).json()
    weather_data.append(response)
    if row_count > 500:
        break
    continue
    
# Get weather data
#weather_response = req.get(endpoint, params=param_dict)
#weather_json = weather_response.json()
#print(weather_json)
# Get temperature from JSON response
#temperature = weather_json["main"]["temp"]
```

```python
# Extract interesting data from responses
lat_data = [data.get("coord").get("lat") for data in weather_data]
temp_data = [data.get("main").get("temp") for data in weather_data]
humidity_data =[data.get("main").get("humidity") for data in weather_data]
cloudiness_data = [data.get("clouds").get("all") for data in weather_data]
wind_speed_data = [data.get("wind").get("speed") for data in weather_data]
weather_data = {"temp": temp_data, "lat": lat_data, "humidity": humidity_data, "cloudiness": cloudiness_data,
               "wind speed": wind_speed_data
               }
weather_data = pd.DataFrame(weather_data)
weather_data.head()
weather_data.to_csv('Weather_Data.csv')
```

```python
plt.scatter(weather_data["lat"],weather_data["temp"], marker ="o", c = "red",
           label = "Temperature(F)")
plt.scatter(weather_data["lat"],weather_data["humidity"], marker ="o", c = "blue",
           label = "Humidity(%)")
plt.scatter(weather_data["lat"],weather_data["cloudiness"], marker ="o", c = "grey",
           label = "Cloudiness(%)")
plt.scatter(weather_data["lat"],weather_data["wind speed"], marker ="o", c = "gold",
           label = "Wind Speed(mph)")
plt.title("Weather Variables V.S. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Weather Variable 12/11/17")
y_limit =104
plt.ylim(-4,y_limit)
plt.grid()
# Now add the legend with some customizations.
legend = plt.legend(loc='best', shadow=True)
plt.savefig("WeatherVSLatitude.png")
plt.show()
```
